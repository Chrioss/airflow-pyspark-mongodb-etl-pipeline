#### Data Validation

```python
# Create a data context

from great_expectations.data_context import FileDataContext

context = FileDataContext.create(project_root_dir='./')
```

```python
datasource_name = "csv-data-Terbaru_M3_Fin"
datasource = context.sources.add_pandas(datasource_name)

asset_name = "Workflow-Terbaru"

asset = datasource.add_dataframe_asset(name=asset_name)

batch_request = asset.build_batch_request(
    dataframe=data_m3
)
```

```python
suite_name = "expectation_data_m3"
context.add_or_update_expectation_suite(expectation_suite_name=suite_name)
validator = context.get_validator(
    batch_request=batch_request,
    expectation_suite_name=suite_name
)
validator.head()
```

#### Expectation


```python
#Ekspektasi 1:  untuk memastikan bahwa nilai Task_ID harus unik
validator.expect_column_values_to_be_unique("Task_ID")
```

```python
#Ekspektasi 1:  untuk memastikan bahwa nilai Workflow_ID harus unik
validator.expect_column_values_to_be_unique("Workflow_ID")
```

```python
#Ekspektasi 2: Kolom Delay_Flag hanya boleh berisi nilai 0(Tidak Delay) atau 1(Delay) pekerjaannya
validator.expect_column_values_to_be_in_set("Delay_Flag", [0,1])
```

```python
#Ekspektasi 3: Kolom Priority_Level hanya boleh berisi nilai Low, Medium, High, atau Critical
validator.expect_column_values_to_be_in_set("Priority_Level", ["Low","Medium","High","Critical"])
```

```python
#Ekspektasi 4: Kolom Approval_Level hanya boleh berisi nilai Level 1, Level 2, atau Level 3
validator.expect_column_values_to_be_in_set("Approval_Level", ["Level 1","Level 2","Level 3"])
```

```python
#Ekspektasi 5: Kolom Estimated_Time_Minutes dan Actual_Time_Minutes hanya boleh berisi nilai antara 0 dan 1440 (24 jam)
validator.expect_column_values_to_be_between("Estimated_Time_Minutes", min_value=0, max_value=1440)
```

```python
validator.expect_column_values_to_be_between("Actual_Time_Minutes", min_value=0)
```

```python
#Ekspektasi 6: Kolom Task_End_Time harus lebih besar dari Task_Start_Time
validator.expect_column_pair_values_A_to_be_greater_than_B(column_A="Task_End_Time",column_B="Task_Start_Time")
```

```python
#Ekspektasi 7: Kolom Employee_Workload harus berisi nilai yang lebih besar dari 0 karena tidak mungkin pekerjaannya tidak ada atau 0, maka harus lebih besar dari 0
validator.expect_column_values_to_be_between("Employee_Workload", min_value=1)
```

```python
#Ekspektasi 8: Kolom Task_ID tidak boleh bernilai null
validator.expect_column_values_to_not_be_null("Task_ID")
```

```python
#Ekspektasi 9: kolom Workflow_ID harus bernilai WF_(Angka)
validator.expect_column_values_to_match_regex("Workflow_ID", r"^WF_\d+$")
```

```python
#Ekspektasi 10: kolom Task_ID harus bernilai TASK_(Angka)
validator.expect_column_values_to_match_regex("Task_ID",r"^TASK_\d+$")
```

```python
validator.get_expectation_suite().expectations
```

```python
validator.save_expectation_suite()
```

```python
#menambahkan checkpoint untuk menjalankan validasi
checkpoint = context.add_or_update_checkpoint(
    name="cp_data_m3_final",
    validations=[{ "batch_request": batch_request,"expectation_suite_name": suite_name,}],
)
```
